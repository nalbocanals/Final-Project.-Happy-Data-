```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
appshealth=pd.read_excel('/Users/nuria/Documents/datasciencecourse/FINAL PROJECT/appshealthdef_columns.xlsx')
```

```python
appshealth.info()
```

```python
#rename the name of the columns
appshealth.rename(columns={'name':'App_name','typw':'App_type','rate':'App_rate',
                          'descarrgas':'App_downloads+','developer':'App_developer',
                          'update':'App_last_update','reviews':'App_review'}, errors="raise",inplace=True)
```

```python
appshealth.info()
```

```python
column_to_get=['App_name','App_type','App_rate','App_downloads+','App_developer','App_last_update','App_review']
Apps_df=appshealth[column_to_get]
```

```python
Apps_df.head()
```

```python
Apps_df.info()
```

##### Step 2: convert the columns with the number of downloads into a numeric category

```python
Apps_df['App_downloads+'].unique()
```

```python
Apps_df['App_downloads+'] = Apps_df.loc[:,'App_downloads+'].str.replace('+', '')
```

```python
Apps_df.info()
```

```python
Apps_df['App_downloads+'] = Apps_df.loc[:,'App_downloads+'].str.replace('.', '')
```

```python
Apps_df['App_downloads+']=Apps_df['App_downloads+'].astype(int)
```

```python
Apps_df
```

```python
Apps_df.info()
```

```python
#Let's add a 'No review' in the column reviews for those that don't have any review
Apps_df['App_review'] = Apps_df['App_review'].fillna(value='No review')
```

```python
Apps_df.info()
```

```python
#save the csv with our clean data
```

```python
Apps_df.to_csv("Apps_Mental_health_def.csv")
```
