##### STEP 7:
   - Take care of the null values in Health_life_exp_age in my final dataset Happyness

```python
#get the countries
countries=Happyness_df.Country
type(countries)
```

```python
avg_age_country=pd.DataFrame(columns=['Country','Avg_age'])
avg_age_country
```

```python
#get the average of the Health_life_expectancy for each country
liiiiist=[]
for item in countries:
    pd_temp=Happyness_df[Happyness_df.Country==item]
    avg_age=pd_temp.Helth_life_exp_age.mean()
    listtt=[item,avg_age]
    liiiiist.append(listtt)
    superdf = pd.DataFrame(listtt,columns = ['ups'])
    superdf =superdf.T
    superdf.rename(columns={0:'Country',1:'Avg_age'}, inplace=True)
    superdf.reset_index(inplace=True)
    pd.concat([avg_age_country, superdf], ignore_index=True)
```

```python
liiiiist
```

```python
Contry_Ageavg = pd.DataFrame(liiiiist, columns=['Country','Age_avg'])
```

```python
Contry_Ageavg
```

- To replace the NaN values in my final dataset I replace the NaN for 0 and then I merge the datset with the countries and age average. Once I have the merge dones I replace the 0 values for the average. 

```python
Happyness_clean_df=Happyness_df.merge(Contry_Ageavg, left_on='Country',right_on='Country',how='left')
```

```python
Happyness_clean_df.drop_duplicates(inplace=True)
```

```python
Happyness_clean_df.fillna(value=0, inplace=True)
```

```python
Happyness_clean_df.loc[Happyness_clean_df['Helth_life_exp_age']==0,'Helth_life_exp_age']=Happyness_clean_df.Age_avg
```
